```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (20,10)
sns.countplot(x='date', data=df)
plt.show()
```

```python
data = df.total.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
```

```python
ax=df['venue'].value_counts().sort_values(ascending=True).plot.barh(width=.9,color=sns.color_palette("inferno",40))
ax.set_xlabel("Count")
ax.set_ylabel("Grounds")
plt.title("Venues played (from most to least)")
plt.show()
```

```python
ax=df['batsman'].value_counts().sort_values(ascending=True).plot.barh(width=.9,color=sns.color_palette("inferno",40))
ax.set_xlabel("Count")
ax.set_ylabel("Grounds")
plt.title("Venues played (from most to least)")
plt.show()
```
